Explainatory data analysis(EDA) Functions

```python
def check_missing_values(df: pd.DataFrame )-> None :
  if df.empty:
    print("DataFrame is empty.")
    return df
  missing_values = df.isnull().sum()
  missing_percentage  = (missing_values / len(df)) * 100

  missing_values_df = pd.DataFrame({ 'Missing Values': missing_values.values , "Missing Percentage": missing_percentage })

  missing_values_df = missing_values_df.sort_values(by = ['Missing Percentage'] , ascending=False)

  '''if df.isnull().sum().sum() != 0 :
     null_columns = df.columns[df.isnull().any()]
    # Creating heatmap for only those columns
     plt.figure(figsize=(12, 6))
     sns.heatmap(df[null_columns].isnull(), cbar=False, cmap='viridis', yticklabels=False)
     plt.title('Heatmap of Missing Values (Only Null Columns)')
     plt.show()'''

  return missing_values_df

def outliers_detection(df: pd.DataFrame, column:str) -> pd.DataFrame:
  if df.empty:
    print("DataFrame is empty.")
  if column not in df.columns :
    raise ValueError(f"Column '{column}' not found in dataFrame ")
  if not 	pd.api.types.is_numeric_dtype(df[column]):
    raise TypeError(f"Column '{column}' is not numerical.")

  outlier_df = df.copy()


  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1

  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

  return outliers

def modify_data_type(df: pd.DataFrame, column: str, new_dtype) -> pd.DataFrame:
  if df.empty:
    print("DataFrame is empty.")
    return df
  if column not in df.columns:
    raise ValueError(f"Column '{column}' not found in DataFrame.")
  df[column] = df[column].astype(new_dtype)
  return df


def get_unique_rows(df: pd.DataFrame , columns = None) -> pd.DataFrame:
  if df.empty:
    print("DataFrame is empty.")
    return df
    # If no columns are specified, use all columns
  if columns is None:
    columns = list(df.columns)
  if not isinstance(columns , list):
    columns = [columns]
  unique_df = df.drop_duplicates(subset=columns)
  return unique_df


def correlation_analysis(df: pd.DataFrame) -> None:
  if df.empty:
    print("DataFrame is empty.")
    return df
  correlation_matrix = df.select_dtypes(include=np.number).corr()
  print(correlation_matrix)


def get_statistcal_summary(df: pd.DataFrame) -> None:
  if df.empty:
    print("DataFrame is empty.")
  numerical_summary = df.describe().T
  print("Numerical Summary:")
  print(numerical_summary)
  print("-"*100)
  if df.select_dtypes(include=['object']).any().any():
    categorical_summary = df.describe(include=['object']).T
  else:
    categorical_summary = None
  if categorical_summary is not None:
    print("-"*100)
    print("Categorical Summary:")
    print(categorical_summary)
  return None

def remove_outliers(df:pd.DataFrame, column: str)-> pd.DataFrame:
    if df.empty:
        print("DataFrame is empty.")
        return df
    if column not in df.columns:
        raise ValueError(f"Column '{column}' not found in DataFrame.")
    if df.empty:
        print("DataFrame is empty.")
        return df

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Keep only rows within the bounds
    df_clean = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    return df_clean


def get_dataframe_information(df: pd.DataFrame) -> None:
  if df.empty:
    print("DataFrame is empty.")
    return
  print("DataFrame Information:")
  print(df.info())
  print("-"*100)
  print("\nDataFrame Shape:")
  print(f"\nDataFrame shape: {df.shape} (Rows: {df.shape[0]}, Columns: {df.shape[1]})")
  print("-"*100)
  print("\n  DataFrame Columns:")
  print(df.columns)
  print("-"*100)
  print("\n DataFrame Object columns:")

  print(df.select_dtypes(include=['object']).columns)
  print("-"*100)
  print("DataFrame Numeric types")
  print(df.select_dtypes(include=['number']).columns)
  print("-"*100)
  print("\nDataFrame Data Types:")
  print(df.dtypes)
  print("-"*100)
  print("\nDataFrame Head:")
  print(df.head())
  print("-"*100)
  print("\nDataFrame Tail:")
  print(df.tail())
  print("-"*100)
  print("\nDataFrame Sample:")
  print(df.sample(5))
  print("-"*100)
  print("\nDataFrame Counts")
  print(df.count())
  print("-"*100)
  return None

def manipulating_missing_values(df: pd.DataFrame, strategy: str = 'mean', fillna=None, columns=None) -> pd.DataFrame:
    if df.empty:
        print("DataFrame is empty.")
        return df

    df_copy = df.copy()

    if strategy == 'drop':
        df_copy.dropna(inplace=True)
        return df_copy

    if columns is None:
        if strategy in ['mean', 'median']:
            impute_cols = df_copy.select_dtypes(include=['number']).columns
        elif strategy == 'most_frequent':
            impute_cols = df_copy.select_dtypes(include=['object', 'category']).columns
        elif strategy == 'constant':
            if fillna is None:
                print("Error: 'fillna' value required for 'constant' strategy.")
                return df_copy
            impute_cols = df_copy.columns
        else:
            print(f"Invalid strategy '{strategy}'.")
            return df_copy
    else:
        # Ensure columns is a list
        if isinstance(columns, str):
            columns = [columns]

        impute_cols = [col for col in columns if col in df_copy.columns and df_copy[col].isnull().any()]
        if not impute_cols:
            print("No specified columns with missing values.")
            return df_copy

    for col in impute_cols:
        if strategy == 'constant':
            df_copy[col] = df_copy[col].fillna(fillna)
        else:
            imputer = SimpleImputer(strategy=strategy)
            df_copy[col] = imputer.fit_transform(df_copy[[col]]).ravel()

    return df_copy

def visualize_outliers(df: pd.DataFrame, column: str) -> None:
  if df.empty:
    print("DataFrame is empty.")
    return
  if column not in df.columns:
    raise ValueError(f"Column '{column}' not found in DataFrame.")
  if not pd.api.types.is_numeric_dtype(df[column]):
    raise TypeError(f"Column '{column}' is not numerical.")
  #boxplot
  plt.figure(figsize=(10, 5))
  sns.boxplot(x=df[column], color='lightblue',showmeans= True , meanprops={'marker':'*', 'markeredgecolor': 'r' })
  plt.title(f'Boxplot of {column}')
  plt.show()
  print("\n")
  print("-"*100)
  return

def visualize_relationships(df:pd.DataFrame , x: str , y: str) -> None:
  if df.empty:
    print("DataFrame is empty.")
    return
  if x not in df.columns:
    raise ValueError(f"Column '{x}' not found in DataFrame.")
  if y not in df.columns:
    raise ValueError(f"Column '{y}' not found in DataFrame.")
  x_is_categorical = df[x].dtype == 'object'
  y_is_categorical = df[y].dtype == 'object'
  if x_is_categorical and y_is_categorical:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x=x, hue=y)
    plt.title("Relationship between Categorical Variables")
    plt.xlabel(x)
    plt.ylabel("Count")
  if x_is_categorical and not y_is_categorical:
    plt.figure(figsize=(10, 5))
    sns.boxplot(data=df, x=x, y=y)
    plt.title("Relationship between Categorical and Numerical Variables")
    plt.xticks(rotation=90)
    plt.xlabel(x)
    plt.ylabel(y)
  if not x_is_categorical and not y_is_categorical:
    plt.figure(figsize=(10, 5))
    sns.scatterplot(data=df, x=x, y=y)
    plt.title("Relationship between Numerical Variables")
    plt.xlabel(x)
    plt.ylabel(y)
  plt.show()
  return

def univariate_analysis(df: pd.DataFrame, column: str) -> None:
  if df.empty:
    print("DataFrame is empty.")
    return
  if column not in df.columns:
    raise ValueError(f"Column '{column}' not found in DataFrame.")
  if column in df.select_dtypes(include=['object']).columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x=column , order=df[column].value_counts().index, palette='pastel')
    plt.title(f' Count Plot Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.show()
  else:
    plt.figure(figsize=(10, 5))
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    sns.histplot(data=df, x=column, kde=True)
    plt.show()

```

```python

def calculate_earthquake_risk_score(row):
    mag = row['mag']
    depth = row['depth']
    mag_score = mag / 10
    depth_score = (700 - depth) / 700
    depth_score = max(depth_score, 0)  # In case depth > 700
    distribution_factor = np.random.uniform(0.9, 1.1)  # Small ±10% random variation

    risk_score = (0.7 * mag_score + 0.3 * depth_score) * distribution_factor

    risk_score = min(max(risk_score, 0), 1)

    return risk_score
```

```python
earthquake_df['risk_score'] = earthquake_df.apply(calculate_earthquake_risk_score, axis=1)
```

```python

def remove_below_threshold(df: pd.DataFrame, column: str, threshold: float = 400) -> pd.DataFrame:

    if df.empty:
        print("DataFrame is empty.")
        return df
    if column not in df.columns:
        raise ValueError(f"Column '{column}' not found in DataFrame.")
    if not pd.api.types.is_numeric_dtype(df[column]):
        print(f"Warning: Column '{column}' is not numerical. Thresholding may not behave as expected.")

    df_filtered = df[df[column] >= threshold].copy()
    return df_filtered

def drop_invalid_rows(df):
    cleaned_df = df[df['lotArea'] >= df['livingArea']].copy()
    return cleaned_df

real_estate_df1_filtered = drop_invalid_rows(real_estate_df1_sorted)
real_estate_df1_filtered = remove_below_threshold(real_estate_df1_filtered, 'price_per_sq_m')


```

```python

def process_bedroom_and_living_space(df: pd.DataFrame) -> pd.DataFrame:

    df_copy = df.copy()

    required_cols = [ 'livingArea']
    if not all(col in df_copy.columns for col in required_cols):
        print(f"Warning: Missing one or more required columns: {required_cols}. Cannot process.")

    df_copy['livingArea'] = pd.to_numeric(df_copy['livingArea'], errors='coerce').fillna(0)

    df_copy.loc[df_copy['livingArea'] == 0, ['total_baths', 'num_bedrooms.x']] = 0
    df_copy.loc[df_copy['num_bedrooms.x'] == 0, 'num_bedrooms_categorical'] = 'Invalid'
    df_copy.loc[df_copy['total_baths'] == 0, 'num_full_baths_categorical'] = 'Invalid'
    df_copy.loc[df_copy['livingArea'] == 0 , 'yearBuilt.x'] = -1
    return df_copy

real_estate_df1_processed = process_bedroom_and_living_space(real_estate_df1_filtered)


print(real_estate_df1_processed[['num_bedrooms.x', 'livingArea']].head())
```

```python
def adjust_price(base_price, risk_score):
    if risk_score < 0.3:
        bias_percentage = np.random.uniform(0.05, 0.15)
    elif risk_score < 0.6:
        bias_percentage = np.random.uniform(0.15, 0.25)
    else:
        bias_percentage = np.random.uniform(0.25, 0.40)

    random_noise = np.random.uniform(-0.02 * base_price, 0.02 * base_price)

    adjusted_price = base_price - (base_price * risk_score * bias_percentage) + random_noise

    return max(adjusted_price, 0)

real_estate_copy['adjusted_price'] = real_estate_copy.apply(
    lambda row: adjust_price(row['price'], row['total_risk_score_norm']),
    axis=1
)

```
